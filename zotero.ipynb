```python
# Collections Tree
from treelib import Node, Tree
from pprint import pprint

sql = "SELECT * FROM collections"
result = c.execute(sql)

# names = list(map(lambda x: x[0], c.description))
# print(names)
# ['collectionID', 'collectionName', 'parentCollectionID', 'clientDateModified', 'libraryID', 'key', 'version', 'synced']

# Make List of Nodes
nodes = []
for row in result:
  nodes.append(row)

# Tree
tree = Tree()

# Root
for node in nodes:
  if node[1] == 'Nutrition & Fitness':
    tree.create_node(node[1], node[0], data=node)
  elif node[3] is None:
    nodes.remove(node)
    
# Let's try to get everything
while True:
  num_nodes = len(nodes)
  # print(num_nodes)
  
  for node in nodes:
    try:
      tree.create_node(node[1], node[0], parent=node[2], data=node)
      nodes.remove(node)
    except:
      pass

  # print(len(nodes))
  if len(nodes) == num_nodes:
    break
  # print()

tree.show()

print(tree.to_json(with_data=True))
```
